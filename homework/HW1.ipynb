####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
sidewalk_cleanings_ha = pd.read_sql("SELECT * FROM sfdata WHERE category='Street and Sidewalk Cleaning' AND neighborhood='Haight Ashbury'", con)
sidewalk_cleanings_ha
```

####8. Which type of complaint was most common? 

```python
most_common_complaint = sidewalk_cleanings_ha = pd.read_sql("SELECT request_type, count(request_type) as 'Number of Requests' FROM sfdata \
                                                             GROUP BY request_type \
                                                             ORDER BY count(request_type) DESC LIMIT 1", con)
most_common_complaint
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
pd.read_sql("SELECT neighborhood FROM (SELECT neighborhood, request_type, max(total) FROM (SELECT neighborhood, request_type, count(request_type) as 'total'\
                           FROM sfdata GROUP BY request_type) GROUP BY neighborhood) WHERE request_type <> 'Street and Sidewalk Cleaning'", con)
```
